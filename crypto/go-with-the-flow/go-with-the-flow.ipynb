## Nicolas's writeup
So, you had the following information: ciphertext, stream cipher and known plaintext.

The encryption is: message XOR stream = ciphertext (OTP in this case)

Given the key length (16), the stream cipher will generate the OTP sequence, since the message length > 16.

First trick, don't focus on the key. Should be impractical to bruteforce it and ... you don't need it because you have some known plaintext.

XOR operator is such that m XOR k = c implies k = c XOR m
-> you can deduce the stream cipher value if you find the correct offset of the known plaintext.

Second trick, known plaintext is 15 bytes.
You can choose any 16th byte at this point, the stream cipher will produce a dozen correct bytes.
Then you can iterate on every byte of the ciphertext and assume it's where the known plaintext is -> generate the corresponding key stream and decrypt the message.

Third trick, only the first few bytes after the known plaintext will be correct as you had to choose a 16th byte to generate the potential key streams.
Check for the computed messages with only printable bytes for the first 10 bytes after the known plaintext.

You now have the correct offset (or multiple depending on how many bytes you decided to validate).
Bruteforce the 16th byte of your generated key stream(s) and check for the flag or a message with only printable characters.

XOR operator is such that m XOR k = c implies k = c XOR m -> you can deduce the stream cipher value if you find the correct offset of the known plaintext.

`````python
def xor(b1, b2):
    return bytes([_a ^ _b for _a, _b in zip(b1, b2)])

def xor_4(a, b, c, d):
    # TODO: Make this function more generic to support any number of params?
    return xor(a, xor(b, xor(c, d)))

test_message = b'hello there :)'
test_key = b'general kenobi'
test_encrypted_message = xor(test_message, test_key)
# Test a xor property: A XOR B = C => C XOR A = B
test_decrypt = xor(test_encrypted_message, test_message)
assert test_decrypt == test_key
`````

You can choose any 16th byte at this point, the stream cipher will produce a dozen correct bytes. Then you can iterate on every byte of the ciphertext and assume itâ€™s where the known plaintext is -> generate the corresponding key stream and decrypt the message.

`````python
def get_keystream(known_keystream_part, known_part_index):
    global key_length, ciphertext
    assert len(known_keystream_part) == key_length

    # Code based on X.A.N.A. implementation (see below)

    # The key_stream contains 1 byte per ciphertext byte (don't know what the +1 is, probably the padding)
    key_stream = [None]*(len(ciphertext)+1)

    # The known_keystream_part determines what the keystream is in the given position
    for j in range(key_length):
        key_stream[known_part_index+j] = known_keystream_part[j]

    # compupte backwards: calculate the contents of the key stream in front of the known part
    for j in range(known_part_index-1, -1, -1):
        key_stream[j] = key_stream[j+1] ^ key_stream[j+2] ^ key_stream[j+3] ^ key_stream[16 + j]
        # key_stream[j] = xor_4(key_stream[j+1], key_stream[j+2], key_stream[j+3], key_stream[16+j])

    # extend forward: calculate the contents of the key stream after the know part
    for j in range(known_part_index, len(key_stream)-key_length):
        # 16 bytes key, first 4 bytes XOR -> new byte in keystream
        key_stream[j+16] = key_stream[j] ^ key_stream[j+1] ^ key_stream[j+2] ^ key_stream[j+3]
        # key_stream[j+16] = xor_4(key_stream[j], key_stream[j+1], key_stream[j+2], key_stream[j+3])

    assert None not in key_stream
    #assert known_keystream_part in key_stream
    return key_stream
`````

`````python
import string

def real_text_score(text):
    # Gives a number determing how realistic the given text is (and not just random bytes)
    alphabet = string.ascii_lowercase + string.ascii_uppercase + " ."
    score = 0
    for byte in text:
        letter = chr(byte)
        if letter in alphabet:
            score += 1
    return score
`````

`````python
last_byte = bytes(chr(0), "ascii")
# possible_known_plaintext = known_plaintext + last_byte

best_position = {"score": -float("inf"), "position": None, "plaintext_guess": None}
for known_plaintext_position in range(len(ciphertext)-len(known_plaintext)):
    corresponding_ciphertext_substr = ciphertext[known_plaintext_position:known_plaintext_position+len(known_plaintext)]
    known_keystream_part = xor(known_plaintext, corresponding_ciphertext_substr) + last_byte
    key_stream = get_keystream(known_keystream_part, known_plaintext_position)
    plaintext_guess = xor(key_stream, ciphertext)
    assert known_plaintext in plaintext_guess

    # The key is 16 bytes and the first 4 are used to generate the next stream cipher. This means that after the crib, there will be 12 characters of correctly decoded text. (12 = 16 bytes key length - 4 first byte xor'ed to generate new character)
    # Check how many of those seem real text (and not just random chars)
    # print(f"{known_plaintext_position}: {plaintext_guess[known_plaintext_position-10:known_plaintext_position+len(known_plaintext)+10]}")
    known_plaintext_end = known_plaintext_position+len(known_plaintext)
    next_chars = plaintext_guess[known_plaintext_end:known_plaintext_end+key_length-4]
    score = real_text_score(next_chars)
    if score > best_position["score"]:
        best_position = {"score": score, "position": known_plaintext_position, "plaintext_guess": plaintext_guess, "next_chars": next_chars}

known_plaintext_position = best_position["position"]
best_position
`````

Position 296 seems correct (when you look at next_chars). Let's brute force the last byte now.

`````python
best_byte = {"score": -float("inf"), "plaintext": None}
for i in range(128):
    corresponding_ciphertext_substr = ciphertext[known_plaintext_position:known_plaintext_position+len(known_plaintext)]
    known_keystream_part = xor(known_plaintext, corresponding_ciphertext_substr) + bytes(chr(i), "ascii")
    key_stream = get_keystream(known_keystream_part, known_plaintext_position)
    plaintext_guess = xor(key_stream, ciphertext)

    score = real_text_score(plaintext_guess)
    if score > best_byte["score"]:
        best_byte = {"score": score, "plaintext": plaintext_guess}

best_byte
`````
